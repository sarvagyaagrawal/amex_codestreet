```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from imblearn.over_sampling import RandomOverSampler
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv("/content/drive/MyDrive/Amex Codestreet/spending.csv")
```

```python
X=df.iloc[:700,:4]
y=df.iloc[:700,4]

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
oversample = RandomOverSampler(random_state=0)

X_train, y_train = oversample.fit_resample(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,roc_auc_score,cohen_kappa_score,confusion_matrix,roc_curve,balanced_accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
```

```python
pipe = Pipeline([('scaler', preprocessing.MinMaxScaler()), ('LR',LinearRegression())])
```

```python

parameters = {'LR__fit_intercept':[True,False],'LR__normalize':[True,False]}
grid_search = GridSearchCV(pipe, parameters, cv=5)
grid_fit = grid_search.fit(X_train, y_train)

best_par = grid_search.best_params_
print('Best params:', best_par) 

best_dt = grid_fit.best_estimator_
print(best_dt.score(X_test,y_test))
best_dt.predict(np.array([[4000,5000,2000,6000]]))
```

```python
pipe_new = Pipeline([('scaler', preprocessing.MinMaxScaler()), ('LR',LinearRegression(fit_intercept=True,normalize=True))])
```

```python
pipe_new.fit(X_train, y_train)
```

```python
pipe_new.predict(np.array([[4000,5000,2000,6000]]))
```

```python
pipe_new.score(X_test, y_test)
```
